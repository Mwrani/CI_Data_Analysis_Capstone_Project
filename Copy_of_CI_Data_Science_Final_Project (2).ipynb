<h2>Part 2 - b - SALARY ANALYSIS</h2>
NOTE: A normal distribution has the following attributes:<br>
68% within 1 standard deviation, 95% within 2 STD, 99% within 3 STD<br>
1 - Decode the salaries column

```python
def find_salary(n):
  result = ''
  salary = str(n).replace('3X', '').replace('$', '')
  salary_key = {0:7,1:8,2:9,3:0,4:1,5:2,6:3,7:4,8:5,9:6}
  for char in salary:
    salaryv = salary_key[int(char)]
    result = str(result) + str(salaryv)
  return int(result)

def cvrt_int(n):
  salary = str(n).replace('$', '')
  return int(salary)

mdf_1 = mdf.copy()

mdf_1['salary_decoded'] = mdf_1['salary'].apply(lambda x: find_salary(x) if str(x).startswith('3X') else cvrt_int(x))
display(mdf_1)
```

2 - Create a histogram of all salaries in deciles

```python
import matplotlib.pyplot as plt
import seaborn as sns

mdf_1['salary_decoded'] = pd.to_numeric(mdf_1['salary_decoded'], errors='coerce')

plot_data = mdf_1.dropna(subset=['salary_decoded'])

plt.figure(figsize=(12, 6))
sns.histplot(plot_data['salary_decoded'], bins=30, kde=True, edgecolor='black')
plt.title('Distribution of Decoded Salaries')
plt.xlabel('Decoded Salary')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

mean_salary = plot_data['salary_decoded'].mean()
plt.axvline(mean_salary, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_salary:.2f}')
plt.legend()

plt.tight_layout()
plt.show()
```

4 - Is the salary distribution a normal distribution?

```python
mean = mdf_1['salary_decoded'].mean()
std_dev = mdf_1['salary_decoded'].std()


for i in range(1,4):
  lower_bound = mean - (std_dev*i)
  upper_bound = mean + (std_dev*i)
  imperical = mdf_1[(mdf_1['salary_decoded'] >= lower_bound) & (mdf_1['salary_decoded'] <= upper_bound)]
  percentage__std = (len(imperical) / len(mdf_1)) * 100
  print(f"The percentage of salaries within {i} standard deviation of the mean is: {percentage__std:.2f}%")
```

This Histogram shows that the salaries have a positive Kurtosis which is also Leptokurtic. According to the imperical rule which is the data should fall within 68% of 1 standard deviation, 95% of 2 standard deviations and 99% within 3 standard deviations. My data shows 70.65%, 94.4%, and 99.71%. This shows the data is a bit more bunched up closer together in the beginning but changes to a more normal distribution as it goes on.

1 - Import the file raises_CAPSTONE.txt

```python
import pandas as pd
fn2 = 'raises_CAPSTONE.txt'

raises_df = pd.read_csv(fn2)

display(raises_df)
```

2 - The rules for giving a raise are in the file but be careful of the order that your run the raise

```python
import pandas as pd
from datetime import datetime

current_date = pd.to_datetime(datetime.now().strftime('%Y-%m-%d'))

msdf1['hiredate'] = pd.to_datetime(msdf1['hiredate'])
msdf1['years_worked'] = ((current_date - msdf1['hiredate']).dt.days / 365.25).astype(int)

display(msdf1)
```

Calculate the total salary for each dept

3 - Based on each employees salary calculate 2 fields 'Raise_Amount' and 'New_Salary'

```python
import numpy as np

raise_df1 = raises_df.copy()
raise_df1['raise_amount_float'] = raise_df1['raise_amount'].str.replace('%', '').str.replace("'", "").astype(float) / 100
raise_df1 = raise_df1.sort_values(by='years', ascending=False)

conditions = [msdf1['years_worked'] >= year for year in raise_df1['years']]
choices = raise_df1['raise_amount_float'].tolist()

msdf1['emp_raise_multiplier'] = np.select(conditions, choices, default=0.0)

msdf1['Raise_Amount'] = (msdf1['emp_raise_multiplier'] * msdf1['salary_decoded']).round().astype(int)
msdf1['New_Salary'] = ((1 + msdf1['emp_raise_multiplier']) * msdf1['salary_decoded']).round().astype(int)

msdf1 = msdf1.drop(columns=['emp_raise_multiplier'])

display(msdf1)
```

4 - Calculate the total salary for each dept

```python
total_salaries_by_dept = msdf1.groupby('deptCode')['New_Salary'].sum()

for dept_code, total_salary in total_salaries_by_dept.items():
    print(f' The total salary for the {dept_code} Department is {total_salary}')
```

5 - Chart the total salary for each dept in 1 bar chart

```python
import matplotlib.pyplot as plt

dsdf = total_salaries_by_dept.reset_index(name='Total Salary')
dsdf = dsdf.sort_values('Total Salary', ascending=True)
plt.figure(figsize=(12, 7))
plt.barh(dsdf['deptCode'], dsdf['Total Salary'], color='skyblue')
plt.xlabel('Total Salary')
plt.ylabel('Department')
plt.title('Total New Salary by Department')
plt.grid(axis='x', linestyle='--', alpha=0.7)

for index, value in enumerate(dsdf['Total Salary']):
    plt.text(value, index, f'${value:,.0f}', va='center', ha='left')

plt.tight_layout()
plt.show()
```

6 - Create a pie chart that shows the percentage that each dept has of the total money allocated for raises

```python
import matplotlib.pyplot as plt

dept_raise_allocations = total_salaries_by_dept
percentages = dept_raise_allocations / dept_raise_allocations.sum() * 100

plt.figure(figsize=(9, 9))
plt.pie(
    percentages,
    labels=percentages.index,
    autopct='%1.1f%%',
    startangle=90,
    pctdistance=0.75
)
plt.title('Percentage of Total Raise Allocation by Department')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

7 - Create a pie chart to show the percent of men vs women for the raise money allocated

```python
import matplotlib.pyplot as plt

gender_raise_allocations = msdf1.groupby('gender')['New_Salary'].sum()
percentages_gender = gender_raise_allocations / gender_raise_allocations.sum() * 100

plt.figure(figsize=(8, 8))
plt.pie(
    percentages_gender,
    labels=percentages_gender.index.map({'F': 'Female', 'M': 'Male'}),
    autopct='%1.1f%%',
    startangle=90,
    pctdistance=0.85
)
plt.title('Percentage of Total Raise Allocation by Gender')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

8 - Create a pie chart to show the percent of men vs women for the raise money allocated by dept

```python
import matplotlib.pyplot as plt


dept_gender_raise_allocations = msdf1.groupby(['deptCode', 'gender'])['New_Salary'].sum().unstack(fill_value=0)

for dept_code, allocations in dept_gender_raise_allocations.iterrows():
    total_dept_allocation = allocations.sum()
    if total_dept_allocation == 0:
        print(f"No raise allocation data for department: {dept_code}")
        continue

    percentages = (allocations / total_dept_allocation) * 100

    plt.figure(figsize=(8, 8))
    plt.pie(
        percentages,
        labels=percentages.index.map({'F': 'Female', 'M': 'Male'}),
        autopct='%1.1f%%',
        startangle=90,
        pctdistance=0.85
    )
    plt.title(f'Percentage of Raise Allocation by Gender in {dept_code} Department')
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
```

9 - Create a dataframe for promotions. If the persons salary excedes the salary max for their salary grade. create a column
called 'Promotion' and add the string 'PROMOTION DUE' otherwise leave blank

```python
import pandas as pd

fn3 = 'salary_grade_CAPSTONE.txt'
sgdf = pd.read_csv(fn3)
sgdf.columns = sgdf.columns.str.replace("'", "").str.strip()
sgdf = sgdf.rename(columns={'grade': 'sg'})

sgdf['sg'] = sgdf['sg'].astype(int)
sgdf['Salary_Max'] = sgdf['end_salary'].str.replace('$', '').str.replace(',', '').fillna(0).astype(int)
salary_max_dict = sgdf.set_index('sg')['Salary_Max'].to_dict()

def calculate_promotion_status(row):
  if pd.notna(row['termdate']) and str(row['termdate']).strip() != '':
    return '-'
  current_sg = row['sg']
  new_salary = row['New_Salary']
  max_salary = salary_max_dict.get(current_sg)
  if max_salary is not None and max_salary > 0 and new_salary > max_salary:
    return 'Promotion Due'
  else:
    return '-'

msdf1['Promotion'] = msdf1.apply(calculate_promotion_status, axis=1)

display(msdf1)
```
